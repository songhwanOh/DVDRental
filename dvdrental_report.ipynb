```python
import pandas as pd
import numpy as np

data1 = pd.read_csv('Q1_results.csv')
data2 = pd.read_csv('Q2_results.csv')
data3 = pd.read_csv('Q3_results.csv')
data4 = pd.read_csv('Q4_results.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize = (30,25))
# Q1 plot 
my_plot = sns.histplot(x='filmtitle', y='rentalcount', hue='rentalrank', data=data1, ax=axs[0,0])
my_plot.set(title='Movies Rented Over 30 Times')
my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=45)
# Q2 plot
sns.histplot(x='categoryname', y='filmtitle', hue='quartile',data=data2, ax=axs[0,1]).set(title='Family Movies and Rental Duration Quartiles')
# Q3 plot
sns.barplot(y='Count', x='Category', hue='standard_quartile', data=data3, ax=axs[1,0]).set(title='Rental Duration Quartiles and Count for Family Movies')
# Q4 plot
sns.barplot( x='rental_month', y='count_rentals', hue='store_id', data=data4, ax=axs[1,1]).set(title='Total Rentals per month for Each Store')
```
